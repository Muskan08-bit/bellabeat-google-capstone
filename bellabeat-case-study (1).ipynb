```python
# change data type of Id and ActivityDate column for further analysis
daily_activity_df['Id'] = daily_activity_df['Id'].astype(str)
daily_activity_df['ActivityDate'] = pd.to_datetime(daily_activity_df['ActivityDate'], format = '%m/%d/%Y')
daily_activity_df.dtypes
```

```python
# convert all column names to lower case for standardization
daily_activity_df.columns = daily_activity_df.columns.str.lower()
daily_activity_df.columns
```

```python
# renaming columns in camel case for standardization
daily_activity_df.rename(columns = {'activitydate': 'activity_date', 'totalsteps': 'total_steps', 'totaldistance': 'total_distance', 'trackerdistance': 'tracker_distance',
       'loggedactivitiesdistance': 'logged_activities_distance', 'veryactivedistance':'very_active_distance',
       'moderatelyactivedistance': 'moderately_active_distance', 'lightactivedistance': 'light_active_distance',
       'sedentaryactivedistance': 'sedentary_active_distance', 'veryactiveminutes': 'very_active_minutes', 'fairlyactiveminutes': 'fairly_active_minutes',
       'lightlyactiveminutes': 'lightly_active_minutes', 'sedentaryminutes': 'sedentary_minutes'}, inplace = True)
daily_activity_df.columns
```

```python
# checking if columns total_distance and tracker_distance are same or not because they seem same through the data
daily_activity_df['diff_distances'] = daily_activity_df['total_distance'] - daily_activity_df['tracker_distance']
daily_activity_df['diff_distances'].value_counts()
```

```python
heartrate_df=pd.read_csv('Bellabeat case study/heartrate_seconds_merged.csv')
heartrate_df
```

The dataset contains 1154681 responses to 3 columns. This dataset has value of heart rate in every 5 seconds of 14 different respondents.

```python
len(heartrate_df['Id'].unique())
```

```python
heartrate_df.info()
```

```python
# change data type of time 
heartrate_df['Time'] = pd.to_datetime(heartrate_df['Time'])
```

```python
# separate column from Time to hour
heartrate_df['hour'] = heartrate_df['Time'].dt.floor('h')
heartrate_df
```

```python
hourly_heartrate = heartrate_df.groupby(['Id', 'hour'])[['Value']].mean().reset_index()
hourly_heartrate['only_hour'] = hourly_heartrate['hour'].dt.hour.astype(int)

hourly_heartrate
```

```python
# dividing heart rate category hourly wise based on heart rate value
import numpy as np
hourly_grouped_data = hourly_heartrate.groupby(['Id', 'only_hour'])[['Value']].mean().reset_index() 
hourly_grouped_data['heart_rate_level'] = np.select(
    [
        hourly_grouped_data.Value >= 100,
        hourly_grouped_data.Value.between(60, 100, inclusive = 'left')
    ],
    [
        'High Rate',
        'Resting Rate'
    ],
    default = 'Low Rate'
)
hourly_grouped_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(12, 6))
sns.histplot(data=hourly_grouped_data, x='only_hour', hue='heart_rate_level', multiple='stack', discrete=True)
plt.xlabel('Hour of the Day')
plt.ylabel('Average Hear Rate(BPM)')
plt.title('Hourly Average Heart Rate')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_grouped_data, x='only_hour', y='Value', marker='o', hue='heart_rate_level')
plt.title('Hourly Average Heart Rate Trend')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Heart Rate(BPM)')
plt.grid(True)
plt.show()
```

## Observation and Insights
1. The falling heart rate between 12 AM and 5 AM is consistent with the sleep schedule of the respondents, reflecting a natural resting condition.
2. The peaks in heart rate at 5 AM and 10 AM reflect heightened physical activity levels, possibly because of morning exercise or gym time. Still, the general average resting heart rate is high, signifying that a majority of people do not lead an active life.
3. Peak evening heart rates at 3 PM and 8 PM can be attributed to running or high-intensity exercise, with another time of increased physical activity
3. Because only a portion of users work out, add personalized workout reminders for less active users to enhance consistency.
4. Utilize AI-powered insights to recommend best times to work out based on the user's historical activity data.
5. Daily movement streaks to encourage an active lifestyle.

```python
hourly_calories_df = pd.read_csv('Bellabeat case study/hourlyCalories_merged.csv')
hourly_calories_df
```

```python
hourly_calories_df.info()
```

```python
# Convert 'ActivityHour' to datetime format
hourly_calories_df['ActivityHour'] = pd.to_datetime(hourly_calories_df['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
# Extract only the hour for better analysis
hourly_calories_df['only_hour'] = hourly_calories_df['ActivityHour'].dt.hour.astype(int)
hourly_calories_df
```

```python
# Group by Id and hour to get the average calories burned per hour
hourly_grouped_calories = hourly_calories_df.groupby(['Id', 'only_hour'])[['Calories']].mean().reset_index()
hourly_grouped_calories.describe()
```

```python
hourly_grouped_calories['calories_burn_level'] = np.select(
    [
        hourly_grouped_calories['Calories'] > 250,  # High burn
        hourly_grouped_calories['Calories'].between(101, 250, inclusive='left')  # Moderate burn
    ],
    [
        'High Burn',
        'Moderate Burn'
    ],
    default='Low Burn'  # Default is Low Burn
)
```

```python
# distribution of calories burned per hour
plt.figure(figsize=(12,6))
sns.histplot(x=hourly_grouped_calories['only_hour'], hue=hourly_grouped_calories['calories_burn_level'], multiple='stack', discrete=True)
plt.xlabel('Hour of the Day')
plt.ylabel('Count of Users')
plt.title('Calories Burn Levels by Hour')
plt.show()
```

## Observation
1. People have high burn calories during 5 AM to 10 AM largely because of intense exercise sessions. Moreover, most people have moderate calorie expenditure during the day from daily activities and normal movements.
2. Stille most of the people have low burn calories during the whole day which signifies a largely sedentary lifestyle, possibly due to extended sitting, office work, or lack of organized physical activity.
3. For users with low daily calorie burn, the product can suggest small, achievable goals like step-count challenges, stretch breaks, or guided movement sessions to promote a more active routine.
4. Position Bellabeat as a complete lifestyle companion, not just a fitness tracker. Marketing can highlight how small, daily movements contribute to long-term health, appealing to working professionals, sedentary users, and casual fitness enthusiasts.

```python
#Average calories burned over the day
plt.figure(figsize=(12,6))
sns.lineplot(data=hourly_grouped_calories, x='only_hour', y='Calories', marker='o', hue='calories_burn_level')
plt.title('Hourly Calories Burned Trend')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Calories Burned')
plt.grid(True)
plt.show()
```

## Observation and Insights
1. Because morning (5 AM) and afternoon (3â€“6 PM) are typically times of natural high activity, Bellabeat can provide custom workout suggestions matching these natural highs to engage to the fullest extent.
2. Low calorie burn remains consistent throughout the day, while moderate calorie burn levels also show a steady trend.
3. With such consistent low burn of calories across most of the day, Bellabeat can include "Move More" gentle reminders, short challenges, or micro-workouts guided by micro-exercises so that users keep moving, especially for prolonged inactive hours.

```python
hourly_steps_df = pd.read_csv('Bellabeat case study/hourlySteps_merged.csv')
hourly_steps_df
```

```python
hourly_steps_df.info()
```

```python
# Convert 'ActivityHour' to datetime format
hourly_steps_df['ActivityHour'] = pd.to_datetime(hourly_steps_df['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')
# Extract only the hour for better analysis
hourly_steps_df['only_hour'] = hourly_steps_df['ActivityHour'].dt.hour.astype(int)
# Group by Id and hour to get the average steps per hour
hourly_steps_grouped=hourly_steps_df.groupby(['Id', 'only_hour'])[['StepTotal']].mean().reset_index()
hourly_steps_grouped
```

```python
hourly_steps_grouped.describe()
```

```python
hourly_steps_grouped['activity_segment'] = np.select(
    [
        hourly_steps_grouped['StepTotal'] > 4000,
        hourly_steps_grouped['StepTotal'].between(2001, 4000, inclusive='left'),   
        hourly_steps_grouped['StepTotal'].between(501, 2000, inclusive='left')
    ],
    [
        'Highly Active',
        'Moderately Active',
        'Lightly Active'
    ],
    default= 'Sedentary'
)
hourly_steps_grouped
```

```python
# Distribution of steps per hour
plt.figure(figsize=(12,6))
sns.histplot(data=hourly_steps_grouped, x='only_hour', hue='activity_segment', multiple='stack', discrete=True)
plt.xlabel('Hour of the Day')
plt.ylabel('Count of Users')
plt.title('Hourly Steps Trend by Activity Level')
plt.show()
```

## Observations and Insights
1. Moderate levels of activity between 5 AM and 10 AM are seen primarily owing to exercise classes, while a large number of individuals are lightly active throughout the day in terms of daily activities and usual movement, as observed from their step counts
2. But, a majority of people are still sedentary for the majority of the day, suggesting a lifestyle characterized by excessive sitting, office work, or lack of physical movement.
3. To encourage more movement, the product can offer customizable activity streaks, reminders to stay active, and posture alerts. This helps users add more movement to their daily routine without needing a structured workout.

```python
# Average steps taken over the day
plt.figure(figsize=(12,6))
sns.lineplot(x='only_hour', y='StepTotal', hue='activity_segment', data=hourly_steps_grouped, marker='o')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Total Steps')
plt.title('Hourly Steps Trend with Activity Level')
plt.grid(True)
plt.show()
```

## Observations and Insights
1. Moderate activity is highest around 10 AM and between 3â€“6 PM, based on step counts. Bellabeat can use these active times to offer personalized movement plans or quick workout suggestions to help users stay active and make the most of their energy.
2. People stay inactive for most of the day, with only a small amount of light activity. This suggests that they donâ€™t have regular movement breaks in their daily routine
3. To help break long periods of inactivity, Bellabeat can add custom movement goals, fun challenges, or smart activity suggestions that encourage users to move more during their least active times.

```python
# Extract date from ActivityHour
hourly_steps_df['only_date'] = hourly_steps_df['ActivityHour'].dt.date
hourly_steps_df
```

```python
# Summing up total steps per user per day
active_days=hourly_steps_df.groupby(['Id', 'only_date'])[['StepTotal']].sum().reset_index()
active_days.describe()
```

```python
# Define activity level conditions
conditions = ([
    (active_days['StepTotal']>0) & (active_days['StepTotal']<=2000),
    (active_days['StepTotal']>2000) & (active_days['StepTotal']<=6000),
    (active_days['StepTotal']>6000) & (active_days['StepTotal']<=8000),
    (active_days['StepTotal']>8000)
])
labels = ['Not Active', 'Active', 'Very Active', 'Highly Active']
active_days['active_level']=np.select(conditions, labels, default='Sedentary')
active_days.head(10)
```

```python
# Count frequency of each activity level
activity_counts = active_days['active_level'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=activity_counts.index,  
            y=(activity_counts/active_days.shape[0]) * 100,  
            palette='viridis')
plt.plot(activity_counts.index, (activity_counts/active_days.shape[0]) * 100, 'o--r', markersize=8)
plt.ylabel('Percentage')
plt.xlabel('Activity Level')
plt.title('Activity Trend with Total Steps Criteria')
plt.ylim(0, max((activity_counts/active_days.shape[0]) * 100) + 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Pie chart for activity proportions
plt.figure(figsize=(8,8))
colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
plt.pie(activity_counts, labels=activity_counts.index, autopct='%1.1f%%', colors=colors, startangle=140, explode=(0.05, 0.05, 0.05, 0.05, 0.05))
plt.title('Proportion of Activity Levels')
plt.show()
```

## Observations and Insights
1. With nearly 40% of users already highly active, Bellabeat can introduce progress tracking, achievement badges, or step-based challenges to keep them engaged and motivated.
2. Since 16.9% are sedentary and 7.7% are completely inactive, Bellabeat can focus on step-count reminders, gentle movement prompts, and personalized daily activity goals to encourage more movement.
3. Users with moderate activity (23.9%) can benefit from custom movement streaks, guided exercises, or habit-building programs to gradually transition toward a more active lifestyle.

```python
sleep_data_df = pd.read_csv('Bellabeat case study/minuteSleep_merged.csv')
sleep_data_df
```

```python
sleep_data_df.info()
```

```python
sleep_data_df['date'] = pd.to_datetime(sleep_data_df['date'], format='%m/%d/%Y %I:%M:%S %p')
sleep_data_df['date'] = sleep_data_df['date'].dt.floor('h')
sleep_data_df.head(10)
```

```python
sleep_data_df['hour'] = sleep_data_df['date'].dt.hour
#  Count occurrences of each sleep stage per hour
sleep_stage_counts = sleep_data_df.groupby(['hour', 'value']).size().unstack(fill_value=0)
sleep_stage_counts=sleep_stage_counts.rename(columns = {1: 'light_sleep', 2:'deep_sleep', 3:'REM_sleep'})
sleep_stage_counts
```

```python
plt.figure(figsize=(12,6))
sleep_stage_counts.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(12,6))
plt.xlabel('Hour of the Day')
plt.ylabel('Count of Sleep Stages')
plt.title('Hourly Distribution of Sleep Stages')
plt.legend(title='Sleep Stage')
plt.show()
```

## Observations and Insights
1. Most users experience light sleep during the night (12 AM â€“ 6 AM). Only a small percentage of users enter deep sleep, and even fewer experience REM sleep, though REM sleep increases around 6 AM. Most people go to bed around 12 AM, indicating a common bedtime pattern.
2. Since deep and REM sleep levels are low, Bellabeat can introduce sleep-enhancing content, such as guided breathing exercises, relaxation music, or bedtime wind-down routines to help users achieve more restorative sleep
3. With many users sleeping late (around 12 AM), Bellabeat can suggest customized sleep schedules, bedtime reminders, or relaxation techniques to encourage earlier, higher-quality rest.
